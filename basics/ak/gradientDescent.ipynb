Lets optimise a simple equation.
9 = x + y

After feeding it to a simplistic neural network, the equation would be:

9 = w1 * x + w2 * y

```python
from random import uniform
from matplotlib import pyplot as plt

# let
x = 3.23
y = 6.21
w1 = -0.68
w2 = 0.7
initial_value = w1*x+w2*y
print("our equation is initialised as follows:")
print(f"f(z) = [{w1=}]*[{x=}] + [{w2=}]*[{y=}]")
print(f"with initial value being: {round(initial_value,2)}")
```

```python
# now the gradients are:
w1_grad = y
w2_grad = x

# and let
step_size = 0.1
step = step_size

# lets begin the gradient descent
new_values = []
for _ in range(50):
    w1 += step * w1_grad
    w2 += step * w2_grad
    new_value = round(w1*x+w2*y, 2)
    new_values.append(new_value)
    if new_value < 9: step = step_size # move up
    else: step = -step_size            # move down
    
plt.plot(new_values, "-*", color="brown", label="optimised values")
plt.axhline(9, linestyle="--", color="green", label="desired value")
plt.legend()
plt.title("gradient misses, as step size is too high")
plt.show()
```

```python
# running the same code block as above, but with step size chanaged
w1_grad = y
w2_grad = x
step_size = 0.0001
step = step_size
new_values = []
for _ in range(50):
    w1 += step * w1_grad
    w2 += step * w2_grad
    new_value = round(w1*x+w2*y, 2)
    new_values.append(new_value)
    if new_value < 9: step = step_size # move up
    else: step = -step_size            # move down
plt.plot(new_values, "-*", color="brown", label="optimised values")
plt.axhline(9, linestyle="--", color="green", label="desired value")
plt.legend()
plt.title("gradient is too small, hence unreachable")
plt.show()
```

```python
# running the same code block as above, but with step size chanaged
w1_grad = y
w2_grad = x
step_size = 0.001
step = step_size
new_values = []
for _ in range(50):
    w1 += step * w1_grad
    w2 += step * w2_grad
    new_value = round(w1*x+w2*y, 2)
    new_values.append(new_value)
    if new_value < 9: step = step_size # move up
    else: step = -step_size            # move down
plt.plot(new_values, "-*", color="brown", label="optimised values")
plt.axhline(9, linestyle="--", color="green", label="desired value")
plt.legend()
plt.title("gradient is smooth and reachable")
plt.show()
```

```python
print("our new weights are as follows:")
print(f"f(z) = [{w1=}]*[{x=}] + [{w2=}]*[{y=}]")
print(f"with new value being: {w1*x+w2*y}")
```
