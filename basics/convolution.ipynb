Convolution, put simply, is a linear operation. It is the cross relation to two signals - keep the first signal the same and reverse the second signal. The output is linear time-invariant. Time invariant means that output signal does not depend on the absolute time, that is, a time delay on the input directly equates to time delay of the output. 

```python
from matplotlib import pyplot as plt
from numpy import convolve
```

```python
figure, axis = plt.subplots(1, 3)
plt.subplots_adjust(hspace=1)

# signal 1
def sig_square(x):
  return 0 if x < 3 or x > 5 else 2
sig1 = [sig_square(x/100) for x in range(1000)]
axis[0].set_title('signal#1')
axis[0].set_xticks(ticks=[]); axis[0].set_yticks(ticks=[])
axis[0].plot(sig1)

# signal 1
def sig_triag(x):
  return 0 if x < 0 or x > 2 else x
sig2 = [sig_triag(x/100) for x in range(200)]
axis[1].set_title('signal#2')
axis[1].set_xticks(ticks=[]); axis[1].set_yticks(ticks=[])
axis[1].plot(sig2)

# Perform convolution
conv = convolve(sig1, sig2, mode='valid') 
conv /= len(sig2) # Normalize
axis[2].set_title('1D convolution')
axis[2].set_xticks(ticks=[]); axis[2].set_yticks(ticks=[])
axis[2].plot(conv)
plt.show()
```
