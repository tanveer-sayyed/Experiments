# Information Theory

It allows us to **quantify** the amount of uncertainty in a probability distribution.

# Sources of uncertainity:
- Inherent stochasticity in the system being modeled
- Incomplete observability (can we observe all of the variables that drive the behavior of the system?)
- Incomplete modeling (to reduce uncertainity did the model discard too much information?)

# Random Variables

A random variable is **just** a description of the states that are possible; it
**must** be coupled with a probability distribution that speciﬁes how likely each of
these states are. A random variable must follow some distribution. A distribution reduces uncertainity.

So when we sample-x we can make inferences about the population i.e. the distribution.

# Probability Distributions

The way we describe probability distributions depends on whether the variables are discrete or continuous.


### Probability Mass Function (P)

a probability distribution over discrete variables. it gives probability of **that** specific state
<br>
P(x = x) = 1 ... indicates x = x is certain
<br>
P(x = x) = 0 ... indicates x = x is impossible

PMF(P) can act on many variables at a time.
P(x = x, y = y ) denotes the probability that x = x and y = y simultaneously. We may also write P(x,y) for brevity.

Properties of PMF:
- domain of P must be the set of all possible states of x (if the domain is [working, eating, sleeping, exercise, bathing] then we cannot ask P(watching Netflix))
- Σ over x∈x of P (x) = 1
- ∀ x ∈ x, 0 ≤ P (x) ≤ 1

### Probability Density Function (p)

density = mass / volume

a probability distribution over continuous state. it does not give probability of a specific state rather an inﬁnitesimal region by p(x)*δx.
<br>
To find the probability that x lies within a region [ a, b ], i.e. the mass, is found by integrating
<br>
∫p(x)*δx between [a,b].
<br>
<br>
Within [a, b], if we assume a uniform distribution of real numbers then :
<br>
u (x ; a,b ) = 1 / (b-a) ... (read as: u parameterized by [a,b])
<br>
We can see that this is nonnegative everywhere. Additionally, it integrates to 1.

### Independence and Conditional Independence

Two **random variables** x and y are independent if their probability distribution
can be expressed as a product of two factors, one involving **only** x and one involving
**only** y.
<br>
x ⊥ y : ∀ x∈x, y∈y , p(x=x, y=y) = p (x=x) * p(y=y)

Two **random variables** x and y are conditionally independent those variables require help of a third variable z. Thus their probability distribution can be expressed as:
<br>
x ⊥ y | z : ∀x∈x, y∈y, z∈z, p(x=x,y=y|z=z) = p(x=x|z=z)* p(y=y|z=z)

### Expectation

Expected value **of** a function(f) **over** a probability distribution(p) is the average value the f takes when a random variable(x) is drawn from p.
 - discrete variables: E<sub>x</sub>[f(x)] = Σ over x∈x on P(x) * f(x)
 - continuous variables: E<sub>x</sub>[f(x)] = ∫p(x)*f(x) dx

### Variance

Variance shows how **values of** a function vary as different values of random variable(x) are **sampled from** its distribution. 
<br>
Var(f(x)) = E[ (f(x) − E [f(x)]) <sup>2</sup> ]

- low variance means x is near its expected value
- square root of variance is standard deviation

### Covariance

Covariance shows how **values of** two functions are **linearly related** to each other when random variables are sampled form their distributions. It includes the scale factor as the distribution can be different.

Cov(f(x), g(y)) = E[  (f(x)−E[f(x)]) * (g(y)−E[g(y)])  ]


 - high absolute values of the covariance mean that the values change very much
and are both far from their respective means at the same time.
 - +ve sign means both take high values simultaneously
 - -ve means one take a high value and the other a lower one
 - correlation is simply the linear relationship with the **scaling factor removed**
 - correlation does not mean causation.


### Independence

Independence is a stronger requirement than **zero covariance**, because independence also excludes nonlinear relationships. It is possible for two variables to be dependent but have zero covariance.

### Likelihood

the likelihood is a measure of how well a model explains observed data, given the parameters. This inturn helps to improve the parameters to get the maxmimum best output possible.

### Bernoulli Distribution

The distribution over a **single** binary random variable. It is controlled by a single parameter φ∈[0, 1]. It has following properties
 - P(x=1) = φ ... thus P(x=0) = 1 − φ
 - P(x=x) = φ<sup>x</sup>  (1 − φ )<sup>1−x</sup>
 - E<sub>x</sub> [x] = φ
 - Var<sub>x</sub>(x) = φ (1 − φ)


```python
import numpy as np
np.random.rand(100)
```

```python
from numpy import arange
from matplotlib import pyplot as plt

# phi is probability of occurance
bernoulli_7 = lambda x: 0.7**x * (1-0.7)**(1-x)
plt.hist([bernoulli_7(x) for x in np.random.rand(10000)], label="phi=0.7", edgecolor="lightblue")
plt.legend(loc="upper right")
plt.title("Bernoulli's")
plt.show()
```

### Gaussian Distribution



When we are uncertain so as to choose a distribution, wihout prior knowledge, then we can always use the Gaussian kernel as it immensly common in real world. It servers as a good approximation. Two important reasons why we choose this:
 1. The central limit theorem shows that the **sum** of many independent random variables is approximately normally distributed. Thus in machine learning when we add many layers and errors get accumulated then the total noise would also be normally distributed. Thus noise(something unpredictable earlier) becomes predictable.
 2. out of all possible probability distributions with the same variance, the normal distribution encodes the maximum amount of uncertainty over the real numbers.

- The two parameters µ∈R and σ∈(0,∞) control the normal distribution.
- E[x] = µ
- µ gives the coordinate of the central peak

### The Dirac Distribution

Sometimes we wish to specify that all of the mass in a probability distribution clusters around a single point. This function is deﬁned such that it is zero-valued everywhere except 0, yet integrates to 1. The Dirac delta function is not an ordinary function that
associates each value x with a real-valued output. It is defined in terms of its properties when integrated.

A common use of the Dirac delta distribution is as a component of an empirical
distribution:



which puts probability mass 1/m on each of the m-points of our training set points. Thus whenever we sample some points from this dataset we can view them as being sampled from a distribution. Another important perspective on the empirical distribution is
that it is the probability density that maximizes the likelihood of the training data.
